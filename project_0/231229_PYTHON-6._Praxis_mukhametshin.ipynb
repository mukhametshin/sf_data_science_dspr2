### Задание №1

#### Найти пересечение без дубликатов

**Компания:** «Авито»

___

Пользователь с клавиатуры (через функцию `input()`) вводит через запятую и пробел две последовательности чисел (некоторые идентификаторы товаров/услуг, просмотренных клиентом 'A' и клиентом 'B' на платформе «Авито»).

Необходимо конвертировать каждый ввод пользователя в отдельный список, состоящий из целых чисел, а также составить новый список, состоящий из элементов пересечения двух исходных. Дубликаты необходимо удалить, а результирующий список — вывести на экран.

**Важно!** Если пересечение пустое, то на экран должен выводиться пустой список.

Если ввод пользователя невозможно конвертировать в список из целых чисел, возникает ошибка. Необходимо отлавливать эту ошибку и выводить на экран фразу `Некорректный ввод`.

Для поиска пересечений в списках напишите функцию `pure_intersection()`, которая будет возвращать список с уникальными элементами пересечения.


**Пример №1:**
```
Ввод:
108, 138, 42, 52, 14
109, 13, 52, 32, 42, 14, 109

Вывод:
```

**Пример №2:**

**Пример №3:**
```

Вывод:

```

**Подсказка:**

Для определения пересечения преобразуйте списки во множества и воспользуйтесь методом `intersection()`.

**Критерии оценивания:**

- Задача решена корректно, выполнены все условия (**2 балла**):
    - каждый ввод пользователя конвертируется в список из идентификаторов;
    - функция для определения пересечения корректно работает для любого ввода пользователя, обозначенного в условиях задачи;
    - решение сопровождено комментариями.
- Задача решена корректно, однако не выполнено одно из условий (**1 балл**):
    - ввод пользователя не конвертируется в список;
    - функция для определения пересечения работает только для ограниченного множества входных данных в рамках условия задачи;
    - решение не сопровождено комментариями.
- Задача решена с серьёзными ошибками/не решена (**0 баллов**): представлен некорректный алгоритм определения пересечения элементов.

````python
input_string1 = input('Введите 1-ую последовательность идентификаторов: ')
input_string2 = input('Введите 2-ую последовательность идентификаторов: ')
# ваш код здесь

#строка ошибки некорректного ввода
error_msg_string = "Некорректный ввод"

def pure_intersection(input_string1, input_string2):
  #преобразуем списки в множества
  def to_set(input_string):
    #разделим строку на элементы, выводе -> в список
    input_list = input_string.split(", ")
    #пробуем преобразовать каждый элемент списка в строку
    try:
      #пройдёмся по списку через map и преобразуем элементы
      #списка в int лямбда-функцией
      result = set(map(lambda x: int(x), input_list))
      return result
    except:
      #если что-то пошло не так
      return error_msg_string

  #инициалзируем пустой список для будущих множеств
  input_sets = []
  #пройдёмся по каждой входной строке и преобразуем
  #её в множество, если получится
  for inp in [input_string1, input_string2]:
    result = to_set(inp)
    #если функция to_set() преобразования выдаёт ошибку
    #то передаём эту ошибку объемлющей функции и возвращаем её
    if result == error_msg_string:
      return error_msg_string
    #иначе добавляем множество в список множеств
    else:
      input_sets.append(result)

  #находим пересечение двух множеств
  inters =  input_sets[0].intersection(input_sets[1])
  #и возвращаем его, преобразовав в список
  return list(inters)

print(pure_intersection(input_string1, input_string2))
````

**Подсказка №2:**

**Критерии оценивания:**

    - программа некорректно обрабатывает присутствие слов во вводе пользователя, что приводит к ошибкам при выполнении программы;
    - решение не сопровождено комментариями.
- Задача решена с серьёзными ошибками/не решена (**0 баллов**): программа некорректно рассчитывает минимум и максимум числовой последовательности.

````python
input_string = input('Введите последовательность чисел: ')
# ваш код здесь
def find_min_max(input_string):
  # заменим запись с десятичными запятыми на точки
  input_string = input_string.replace(",", ".")
  # преобразуем str в список по пробелам
  input_list = input_string.split(" ")
  # заменим все буквосодержащие элементы списка на None
  input_list_cleaned = list(map(lambda x: float(x) if not x.isalpha() else None, input_list))
  # удалим все None из предыдущего списка. Это будет финальный список.
  while None in input_list_cleaned:
    input_list_cleaned.remove(None)

  # образуем строку на вывод, здесь же посчитаем макс. числа списка
  output_string = "Минимум: " + str(min(input_list_cleaned)) + "\nМаксимум: " + str(max(input_list_cleaned))
  # возвращаем финальную строку
  return output_string

print(find_min_max(input_string))
````

#### Найти медиану

**Компания:** «Авито»

---
Медиана, как и среднее, является одной из важнейших характеристик для числовых данных в статистике.

**Медиана** — это число, которое находится в середине упорядоченной по возрастанию числовой последовательности.

Например, в последовательности чисел `1, 5, 2, 3, 6` медианой является число `3`, так как если упорядочить числа по возрастанию (`1, 2, 3, 5, 6`), центральным элементом в последовательности будет именно `3`.

**Примечание.** *Если в выборке чётное число элементов, то за медиану принимается среднее между центральными элементами (после упорядочивания).*

Например, в последовательности чисел `100, 5, 2, 4, 3, 6` медианой является число `4.5`, так как, если упорядочить числа по возрастанию (`2, 3, 4, 5, 6, 100`), то центральным элементом в последовательности будет среднее арифметическое между 4 и 5 ($\frac{4 + 5} {2} = 4.5$).

---
Пользователь с клавиатуры (через функцию `input()`) вводит через запятую и пробел неупорядоченную последовательность чисел.

Необходимо конвертировать ввод пользователя в список действительных чисел и вывести на экран медиану этой числовой последовательности.

Для поиска медианы в списке напишите функцию `find_median()`, которая будет возвращать одно число — медианное значение. Результат работы функции нужно вывести на экран.

Если переданная в функцию последовательность окажется пустой (пользователь ничего не ввёл) или пользователь ввёл числа некорректно (ввод нельзя преобразовать в числа), на экран должна быть выведена фраза `Некорректный ввод`.

**Пример №1:**

```

```


**Критерии оценивания:**

- Задача решена корректно, выполнены все условия (**2 балла**):
    - ввод пользователя конвертируется в список из чисел;
    - функция для расчёта медианы работает корректно для списка чётной и нечётной длины;
    - в случае некорректного ввода функция возвращает `None`;
    - решение сопровождено комментариями.
- Задача решена корректно, однако не выполнено одно из условий (**1 балл**):
    - ввод пользователя не конвертируется в список;
    - функция для расчёта медианы некорректно работает для списка чётной или нечётной длины;
    - в случае некорректного ввода функция отрабатывает с ошибкой;
    - решение не сопровождено комментариями.
- Задача решена с серьёзными ошибками/не решена (**0 баллов**): представлен некорректный алгоритм расчёта медианы для обоих случаев, результат работы программы зависит от вводимой последовательности чисел и т. д.

````python
input_string = input('Введите последовательность чисел: ')
# ваш код здесь
def find_median(input_string):
  # если пользователь ввёл пустую строку, завершаем, возвращая ошибку
  if input_string == "":
    return "Некорректный ввод"
  # разделим строку на список по запятой с пробелом
  input_list = input_string.split(", ")

  # пробуем преобразовать каждый элемент списка в строку
  try:
    # пройдёмся по списку через map и преобразуем элементы
    #списка в int лямбда-функцией
    input_list_cleaned = list(map(lambda x: int(x), input_list))
  #если что-то пошло не так
  except:
    # возвратим ошибку
    return "Некорректный ввод"

  # отсортируем список в качестве подготовки к поиску медианы
  input_list_cleaned.sort()

  # найдём длину списка
  list_length = len(input_list_cleaned)

  # если длина списка -- чётное число, то
  if list_length % 2 == 0:
    # подсчитаем медиану как среднее между двумя центральными элементами
    output_val = (input_list_cleaned[list_length // 2 - 1] + input_list_cleaned[list_length // 2]) / 2
    # и возвратим её
    return "Median: " + str(output_val)
  # если длина списка -- нечётное, то
  else:
    # медиана будет стоять ровно посередине списка
    output_val = float(input_list_cleaned[list_length // 2])
    # возвращаем её
    return "Median: " + str(output_val)

find_median(input_string)
````

#### Разложить факториал на простые множители

**Компания:** «СБЕР»

**Уровень:** Сложный

___

Пользователь с клавиатуры (через функцию `input()`) вводит некоторое целое число `n`.

---

Напомним, что **факториал числа $n$ ($n!$)** — это произведение всех натуральных чисел от $1$ до $n$ включительно.

Например, факториал числа $3$ будет равен: $3! = 1 * 2 * 3 = 6$.

---

**Простое число** — число, которое делится (нацело) только на себя и на 1. Например, $13$ — простое число, так как оно делится только на себя (на $13$) и на $1$, а $14$ не является простым числом, так как оно делится на $7$ и на $2$.

Подробнее о простых числах можно почитать [здесь](https://dzen.ru/a/XqFMe_VWclT8NTss).

---
**Простые множители** — это простые числа (не считая 1), на которые число делится без остатка.

Например, для числа $10$ простыми множителями являются числа $5$ и $2$. Для числа $28$ множителями являются числа $7$ и $4$, но простыми множителями будут числа $7$ и $2$, так как $4$ — это $2 * 2$, а число $28$ представляется в виде $28 = 7 * 2^2$.

----

В результате своей работы функция должна возвращать строку вида (приведён пример для `n=12`):

`12! = 2^10 * 3^5 * 5^2 * 7 * 11`

Здесь число слева от ^ — множитель, число справа от ^ — его степень. При этом если множитель встречается один раз, знак степени не выводится. Вывод множителей должен производиться в порядке их возрастания.

**Пример №1:**
```
Ввод:
5

Вывод:
5! = 2^3 * 3 * 5
```

**Пример №2:**
```
Ввод:
22

Вывод:
22! = 2^19 * 3^9 * 5^4 * 7^3 * 11^2 * 13 * 17 * 19
```

**Пример №3:**
```
Ввод:
25

Вывод:
25! = 2^22 * 3^10 * 5^6 * 7^3 * 11^2 * 13 * 17 * 19 * 23

```

**Подсказка №2:**

Алгоритм нахождения простых множителей числа вы можете найти [здесь](https://foxford.ru/wiki/informatika/razlozhenie-chisla-na-mnozhiteli-v-python).

````python
n = input('Введите число: ')
# ваш код здесь
# будем использовать две функции из библиотеки math
from math import factorial, sqrt

# явно преобразуем в int строку с введённым числом
n = int(n)

# найдём его факториал
f = factorial(n)

# ищем простые множители
def factors(n, f):

  # внутренняя функция проверки числа на простоту
  def check_prime(f):
    # инициализируем счётчик множителей
    counter_factors = 0
    # пробуем делить на числа от 1 до самого числа f
    for i in range(1, f+1):
      # увеличим счётчик, если f делится на число без остатка
      if f % i == 0:
        counter_factors += 1
    # если всего 2 делителя (или если число f - это 1 или 2), то число - простое
    if counter_factors == 2 or f == 1 or f == 2:
      return True
    else:
      return False

  # инициализируем словарь для множителей
  factors_dict = {}

  # проходим числа i от 2 до корня из f
  for i in range(2, int(sqrt(f))):
    # инициализируем счётчик множителей
    counter_factors = 0
    # повторяем пока f делится на i нацело
    while f % i == 0:
      # если делится нацело и число i - простое
      if f % i == 0 and check_prime(i):
        # увеличим счётчик множителей
        counter_factors += 1
      # поделим f на i, это будет новое число f
      f /= i
      # добавим i в словарь как ключ со значением - сколько получилось раз
      # на него поделить
      factors_dict[i] = counter_factors

  # формируем строку на вывод
  str_list = []
  # пройдёмся по собранному словарю простых множителей
  for key, val in factors_dict.items():
    s1 = str(key)
    # общий случай - выведем множитель и его степень
    if val != 1:
      s2 = "^" + str(val)
    # специальный случай - если множитель встретился единожды - не надо
    # выводить его степень
    else:
      s2 = ""
    # соединим множитель и его степень в строку и добавим её в список на вывод
    str_list.append(s1 + s2)
  # из списка на вывод сделаем строку на вывод, соединив элементы знаком
  # умножения.
  output_string = " * ".join(str_list)
  # возвратим результат в виде "n! = x^a * y^b * ..."
  return str(n) + "! = " + output_string

print(factors(n, f))
````

#### Заменить дублирующиеся буквы

**Компания:** «СБЕР»

**Уровень:** Сложный

____

Пользователь с клавиатуры (через функцию `input()`) вводит строку, состоящую из символов латинского алфавита.

Необходимо последовательно заменять две одинаковые буквы во введённой строке на следующую по алфавиту букву (важно: заменяем по 2 одинаковых буквы за одну итерацию, не более). Операция замены должна повторяться до тех пор, пока в строке не останутся только уникальные символы.

**Несколько важных моментов:**

* буквы не обязаны быть смежными (смежные — стоящие рядом; в строке `'abc'` буквы a и b, а также b и c являются смежными, в то время как a и c — нет);
* zz заменяем на a;
* вводимая строка состоит из букв латинского алфавита в нижнем регистре.

Напишите функцию `replace_duplicates()`, которая принимает на вход произвольную строку `s`, значение которой вводится с клавиатуры. Функция должна последовательно производить описанную операцию и возвращать строку с оставшимися буквами в любом порядке.

**Пример:**
```
input:
zzzab

step 1: zaba
step 2: zbb
step 3: zc

output:
cz
```

```
input:
hhakafh

step 1: haakfi
step 2: hkfib

output:
hkfib
```

**Подсказка:**

При решении задачи вам может понадобиться латинский алфавит. Вы можете получить его из встроенной в Python библиотеки `string`:

````python
import string
alphabet = string.ascii_lowercase
print(alphabet)
````

````python
# ваш код здесь
import string
alphabet = string.ascii_lowercase

# Список из букв алфавита, нужен для удобного обращения по индексам
alphabet_lst = list(alphabet)

s = input('Введите строку: ')

def replace_duplicates(s):

  # внутренняя функция проверки на дубликаты букв
  def check_dups(s):
    # сравниваем длины множества (дубликаты удалены) и списка (дубликаты м.б.)
    if len(set(list(s))) == len(list(s)):
      # если нет дубликатов, возвратим False, иначе True
      return False
    else:
      return True
  # счётчик шагов
  counter_steps = 0
  # выполняется пока в строке есть дубликаты
  while check_dups(s):
    # проходим по каждому элементу строки
    for chr in s:
      # если есть как минимум два включения
      if s.count(chr) >= 2:
        # уберём оба включения
        temp_lst = list(s)
        temp_lst.remove(chr)
        temp_lst.remove(chr)
        # добавим вместо дупликатов букву, следующую по порядку
        # специальный случай: zz -> a
        if chr == "z":
          temp_lst.append("a")
        # общий случай
        else:
          # добавим новую букву к временному списку,
          # работаем с индексами из списка букв алфавита
          temp_lst.append(alphabet_lst[alphabet_lst.index(chr) + 1])
        # пополняем счётчик шагов
        counter_steps += 1
        # образуем новую, модифицированную строку из временного списка
        s = "".join(temp_lst)
        # и напечатаем её
        print(f"step {counter_steps}: {s}")

replace_duplicates(s)
````
